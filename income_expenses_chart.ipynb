```python
import pandas as pd

df= pd.read_csv('deepseek_csv_20251104_519e60.csv.txt')
```

```python
print(df.head(5))
```

```python
df.dtypes
```

```python
df['category']= df['category'].astype(str).str.strip().map({'ΕΣΟΔΑ':1,'ΕΞΟΔΑ':0})
print(df.head(25))
```

```python
df.dtypes
```

```python
income=0
expenses=0
for index, row in df.iterrows():
  category= row["category"]
  amount= row["amount"]
  if category==0:
    income +=amount
  elif category== 1:
    expenses+=amount
print(income,expenses)
```

```python
chart_values=[income,expenses]
print(chart_values)
```

```python
import matplotlib.pyplot as plt
labels = ['income', 'expenses']

data = pd.Series(chart_values, index=labels)

data.plot.pie(autopct='%1.1f%%', startangle=90, figsize=(6, 6))
plt.title('income/expenses chart')
plt.ylabel('')
plt.show()
```
